# Question 1

```python
df_a=duckdb.query("SELECT Model_Name,Brand,Source,Country,count(case when feature_1=-1 then 1 end) as Feature_1_Negative_Total, count(case when feature_1 = 0 then 1 end) as Feature_1_Neutral_Total,count(case when feature_1 = 1 then 1 end) as Feature_1_Positive_Total,count(case when feature_2=-1 then 1 end) as Feature_2_Negative_Total, count(case when feature_2 = 0 then 1 end) as Feature_2_Neutral_Total,count(case when feature_2 = 1 then 1 end) as Feature_2_Positive_Total,count(case when feature_3=-1 then 1 end) as Feature_3_Negative_Total, count(case when feature_3 = 0 then 1 end) as Feature_3_Neutral_Total,count(case when feature_3 = 1 then 1 end) as Feature_3_Positive_Total,count(case when feature_4=-1 then 1 end) as Feature_4_Negative_Total, count(case when feature_4 = 0 then 1 end) as Feature_4_Neutral_Total,count(case when feature_4 = 1 then 1 end) as Feature_4_Positive_Total FROM df  GROUP BY Model_Name, Brand,Source,Country order by Model_Name,Brand,Source,Country").to_df()
df_a
```

```python
# Rename columns if needed
df_a = df_a.rename(columns={
    'Model_Name': 'Model Name'
})
```

```python
df_a.head()
```

```python
# Convert the DataFrame to an Excel file
output_file = 'Transform_1.xlsx'
df_a.to_excel(output_file, index=False, sheet_name='Sheet1')
```

# Question 2

```python
# Define the data
data = {
    'Topic': ['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4']
}

# Create the DataFrame
df_2 = pd.DataFrame(data)
df_2.head()
```

```python
cross_joined_df = duckdb.query("""SELECT Model_Name,Brand,Source,Country,Topic,count(case when feature_1 = -1 then 1 end) as Negative,count(case when feature_1 = 0 then 1 end) as Neutral,count(case when feature_1 = 1 then 1 end) as Positive FROM df CROSS JOIN df_2 GROUP BY Model_Name,Brand,Source,Country,Topic order by Topic,Model_Name,Brand,Source,Country""").to_df()

cross_joined_df
```

```python
# Convert the DataFrame to an Excel file
output_file = 'Transform_2.xlsx'
cross_joined_df.to_excel(output_file, index=False, sheet_name='Sheet1')
```
